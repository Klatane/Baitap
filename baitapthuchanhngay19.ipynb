2. Hãy xây dựng một hàm có tên là my_function có tham số đầu vào là x. Phương pháp tính toán của hàm như sau:
 (i) Nếu x >0  thì trả về log10(x); (ii) Nếu x <= 0  thì trả về $e^{x}$

```python
def my_function(x):
  if x > 0:
    kq= np.log10(x)
  else:
    kq = np.exp**x
  return kq
n = int(input('nhap n: '))
b = my_function(n)
print(b)
```
